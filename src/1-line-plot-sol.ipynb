```python
import matplotlib.pyplot as plt
import numpy as np

# Task: Create data for the line plot
x = np.arange(0, 11)  # x values from 0 to 10 (inclusive)
y = x**2              # y values as the square of x

# Create line plot
plt.figure(figsize=(8, 6))  # Set figure size (optional for better visibility)
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)  # Line plot with basic styling

# Customize the plot
plt.title('Plot of x vs x²', fontsize=14)  # Add title
plt.xlabel('x values', fontsize=12)        # Label for x-axis
plt.ylabel('y = x²', fontsize=12)          # Label for y-axis
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid with dashed lines

# Display the plot
plt.show()
```
