```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create histogram
plt.figure(figsize=(10, 6))  # Set figure size for clarity
plt.hist(data, bins=range(1, 7), align='left', rwidth=0.8, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Distribution of Numbers', fontsize=14, pad=15)  # Title as required
plt.xlabel('Value', fontsize=12)                          # X-axis label as required
plt.ylabel('Frequency', fontsize=12)                      # Optional but helpful

# Set x-axis ticks
plt.xticks(range(1, 6))

# Add grid for better readability (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
