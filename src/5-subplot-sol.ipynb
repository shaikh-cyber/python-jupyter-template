```python
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
x = np.linspace(0, 10, 100)  # Data for line plot
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']  # Data for bar and pie charts
values = [23, 45, 56, 78]
scatter_x = np.random.rand(50)  # Data for scatter plot
scatter_y = np.random.rand(50)

# Create a figure with 2x2 subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# Line plot (top-left)
ax1.plot(x, y, color='blue', linewidth=2)
ax1.set_title('Line Plot', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Bar plot (top-right)
ax2.bar(categories, values, color=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
ax2.set_title('Bar Chart', fontsize=12)
ax2.set_xlabel('Categories')
ax2.set_ylabel('Values')

# Scatter plot (bottom-left)
ax3.scatter(scatter_x, scatter_y, color='purple', s=50)
ax3.set_title('Scatter Plot', fontsize=12)
ax3.grid(True, linestyle='--', alpha=0.7)
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')

# Pie chart (bottom-right)
ax4.pie(values, labels=categories, autopct='%1.1f%%', startangle=90,
        colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
ax4.set_title('Pie Chart', fontsize=12)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
