```python
import matplotlib.pyplot as plt
import numpy as np

# Create data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Larger figure size for clarity
plt.scatter(x, y, color='blue', s=50)  # Added point size for visibility

# Add annotations for each point
for i, (x_val, y_val) in enumerate(zip(x, y)):
    plt.annotate(f'({x_val}, {y_val})', 
                 (x_val, y_val), 
                 xytext=(5, 5),  # Offset annotation slightly above and to the right
                 textcoords='offset points', 
                 fontsize=10)

# Customize the plot
plt.title('Scatter Plot with Annotations', fontsize=14)  # Specified title
plt.xlabel('X-axis', fontsize=12)  # Optional but helpful
plt.ylabel('Y-axis', fontsize=12)  # Optional but helpful
plt.grid(True, linestyle='--', alpha=0.7)  # Optional grid

# Show the plot
plt.show()
```
