```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create pie chart
plt.figure(figsize=(10, 8))  # Larger figure size for clarity
plt.pie(sizes, 
        labels=labels, 
        autopct='%1.1f%%',  # Display percentages with 1 decimal place
        startangle=90,      # Rotate for better visual layout
        textprops={'fontsize': 12})  # Adjust font size of percentages and labels

# Add title
plt.title('Programming Languages Distribution', fontsize=14)

# Ensure the pie chart is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```
