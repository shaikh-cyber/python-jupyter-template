```python
import matplotlib.pyplot as plt

# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create bar plot with custom colors
plt.figure(figsize=(8, 6))  # Set figure size for better visibility
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']  # Distinct colors for each bar
plt.bar(categories, values, color=colors)

# Add title
plt.title('Bar Plot with Custom Colors', fontsize=14)

# Add value labels on top of each bar
for i, v in enumerate(values):
    plt.text(i, v + 0.1, str(v), ha='center', fontsize=12)

# Optional enhancements: Add axis labels for clarity
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Display the plot
plt.show()
```
